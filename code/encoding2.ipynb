```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
from category_encoders import TargetEncoder
# from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

# Label

```python
df = pd.read_csv('../datasets/cleaned(3).csv')

df_label = df.copy()

label_enc = LabelEncoder()
df_label["Accident_severity"] = label_enc.fit_transform(df_label["Accident_severity"])

ordinal_features = df_label.select_dtypes(include=['object', 'category']).columns
ordinal_enc = OrdinalEncoder()
df_label[ordinal_features] = ordinal_enc.fit_transform(df_label[ordinal_features])
```

```python
df_label
```

```python
X_label = df_label.drop(columns=["Accident_severity"])
y_label = df_label["Accident_severity"]

model = RandomForestClassifier(n_estimators=100, random_state=42)

scores = cross_val_score(model, X_label, y_label, cv=5, scoring="accuracy")

print("Mean Accuracy:", scores.mean())
print("Standard Deviation:", scores.std())
```

# OHE

```python
df = pd.read_csv('../datasets/cleaned(3).csv')

# Salin DataFrame untuk modifikasi
df_ohe = df.copy()

# Label Encoding untuk kolom tertentu
label_enc = LabelEncoder()
df_ohe["Accident_severity"] = label_enc.fit_transform(df_ohe["Accident_severity"])

# Pilih kolom yang akan di-OHE-kan (misalnya kolom bertipe 'object' atau 'category' selain 'Accident_severity')
exclude_cols = ['Accident_severity']  # Kolom yang ingin dikecualikan
one_hot_features = df_ohe.select_dtypes(include=['object', 'category']).columns.difference(exclude_cols)

# Terapkan OHE hanya pada kolom yang dipilih
df_ohe = pd.get_dummies(df_ohe, columns=one_hot_features, drop_first=True)

print(df_ohe.head())
```

```python
X_Ohe = df_ohe.drop(columns=["Accident_severity"])
y_ohe = df_ohe["Accident_severity"]

model = RandomForestClassifier(n_estimators=100, random_state=42)

scores = cross_val_score(model, X_Ohe, y_ohe, cv=5, scoring="accuracy")

print("Mean Accuracy:", scores.mean())
print("Standard Deviation:", scores.std())
```

```python


df = pd.read_csv('../datasets/cleanedfix.csv')

df_encoded = df.copy()

label_enc = LabelEncoder()
df_encoded["Accident_severity"] = label_enc.fit_transform(df_encoded["Accident_severity"])

ordinal_features = ['Age_band_of_driver', 'Driving_experience']
ordinal_enc = OrdinalEncoder()
df_encoded[ordinal_features] = ordinal_enc.fit_transform(df_encoded[ordinal_features])

one_hot_features = [
    'Sex_of_driver', 'Vehicle_driver_relation', 'Type_of_vehicle', 'Owner_of_vehicle',
    'Area_accident_occured', 'Lanes_or_Medians', 'Road_allignment', 'Types_of_Junction',
    'Road_surface_type', 'Road_surface_conditions', 'Light_conditions', 
    'Weather_conditions', 'Vehicle_movement','Time'
]
df_encoded = pd.get_dummies(df_encoded, columns=one_hot_features, drop_first=True)

target_features = ['Cause_of_accident', 'Type_of_collision']
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for feature in target_features:
    df_encoded[feature + "_kfold_enc"] = 0 
    
    for train_idx, val_idx in kf.split(df_encoded):
        train_mean = df_encoded.iloc[train_idx].groupby(feature)['Accident_severity'].mean()
        smooth_value = (train_mean * 5 + df_encoded["Accident_severity"].mean()) / (5 + 1) 
    
        df_encoded.iloc[val_idx, df_encoded.columns.get_loc(feature + "_kfold_enc")] = df_encoded.iloc[val_idx][feature].map(smooth_value)

# Hapus fitur asli setelah encoding
df_encoded.drop(columns=target_features, inplace=True)

# Hapus fitur asli setelah encoding
# df_encoded.drop(columns=target_features, inplace=True)
```
