# Target

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import category_encoders as ce
from sklearn.preprocessing import LabelEncoder

# Load the data
df = pd.read_csv('../datasets/cleaned(3).csv')

# Make a copy for target variable processing
df_target = df.copy()

# Label encode the 'Accident_severity' column
label_enc = LabelEncoder()
df_target["Accident_severity"] = label_enc.fit_transform(df_target["Accident_severity"])

# Split into X (features) and y (target)
X_Target = df_target.drop(columns=["Accident_severity"])
y_Target = df_target["Accident_severity"]

# Identify categorical features for encoding
exclude_cols = []  # Define columns you want to exclude from encoding (e.g., already numeric columns)
target_features = X_Target.select_dtypes(include=['object', 'category']).columns.difference(exclude_cols)

# Select the categorical features for encoding
X = X_Target[target_features].astype(str)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_Target, test_size=0.2, random_state=42)

# Initialize TargetEncoder with the correct parameters
encoder = ce.TargetEncoder(smoothing=0.1)

# Fit and transform the encoder on the training data
X_train_encoded = encoder.fit_transform(X_train, y_train)
X_test_encoded = encoder.transform(X_test)

# Initialize RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model on the encoded training data
model.fit(X_train_encoded, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_encoded)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the RandomForest model:", accuracy)

# You can also get the mean accuracy with cross-validation if desired
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(model, X_train_encoded, y_train, cv=5, scoring='accuracy')

print("Mean Accuracy (CV):", cv_scores.mean())
print("Standard Deviation (CV):", cv_scores.std())
```

# KKfold Target

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import category_encoders as ce

# Load the data
df = pd.read_csv('../datasets/cleaned(3).csv')

# Make a copy for target variable processing
df_target = df.copy()

# Label encode the 'Accident_severity' column
label_enc = LabelEncoder()
df_target["Accident_severity"] = label_enc.fit_transform(df_target["Accident_severity"])

# Split into X (features) and y (target)
X_Target = df_target.drop(columns=["Accident_severity"])
y_Target = df_target["Accident_severity"]

# Identify categorical features for encoding
exclude_cols = []  # Define columns you want to exclude from encoding (e.g., already numeric columns)
target_features = X_Target.select_dtypes(include=['object', 'category']).columns.difference(exclude_cols)

# Select the categorical features for encoding
X = X_Target[target_features].astype(str)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_Target, test_size=0.2, random_state=42)

# Initialize KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize TargetEncoder
encoder = ce.TargetEncoder(smoothing=0.1)

# Create an empty array to hold encoded values
X_train_encoded = X_train.copy()

# KFold Cross-validation Target Encoding
for train_idx, val_idx in kf.split(X_train):
    # Split the training data into current fold's train and validation
    X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    # Fit TargetEncoder on the training fold and transform both train and validation fold
    encoder.fit(X_train_fold, y_train_fold)
    X_train_encoded.iloc[val_idx] = encoder.transform(X_val_fold)
    
# Initialize RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model on the encoded training data
model.fit(X_train_encoded, y_train)

# Make predictions on the test set (which needs encoding too)
X_test_encoded = encoder.transform(X_test)
y_pred = model.predict(X_test_encoded)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the RandomForest model with KFold target encoding:", accuracy)

# You can also get the mean accuracy with cross-validation if desired
cv_scores = cross_val_score(model, X_train_encoded, y_train, cv=5, scoring='accuracy')

print("Mean Accuracy (CV):", cv_scores.mean())
print("Standard Deviation (CV):", cv_scores.std())
```

# MIX

```python


df = pd.read_csv('../datasets/cleanedfix.csv')

df_encoded = df.copy()

label_enc = LabelEncoder()
df_encoded["Accident_severity"] = label_enc.fit_transform(df_encoded["Accident_severity"])

ordinal_features = ['Age_band_of_driver', 'Driving_experience']
ordinal_enc = OrdinalEncoder()
df_encoded[ordinal_features] = ordinal_enc.fit_transform(df_encoded[ordinal_features])

one_hot_features = [
    'Sex_of_driver', 'Vehicle_driver_relation', 'Type_of_vehicle', 'Owner_of_vehicle',
    'Area_accident_occured', 'Lanes_or_Medians', 'Road_allignment', 'Types_of_Junction',
    'Road_surface_type', 'Road_surface_conditions', 'Light_conditions', 
    'Weather_conditions', 'Vehicle_movement','Time'
]
df_encoded = pd.get_dummies(df_encoded, columns=one_hot_features, drop_first=True)

target_features = ['Cause_of_accident', 'Type_of_collision']
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for feature in target_features:
    df_encoded[feature + "_kfold_enc"] = 0 
    
    for train_idx, val_idx in kf.split(df_encoded):
        train_mean = df_encoded.iloc[train_idx].groupby(feature)['Accident_severity'].mean()
        smooth_value = (train_mean * 5 + df_encoded["Accident_severity"].mean()) / (5 + 1) 
    
        df_encoded.iloc[val_idx, df_encoded.columns.get_loc(feature + "_kfold_enc")] = df_encoded.iloc[val_idx][feature].map(smooth_value)

# Hapus fitur asli setelah encoding
df_encoded.drop(columns=target_features, inplace=True)

# Hapus fitur asli setelah encoding
# df_encoded.drop(columns=target_features, inplace=True)
```
