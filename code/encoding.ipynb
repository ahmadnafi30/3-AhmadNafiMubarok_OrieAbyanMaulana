```python

label_encode_features = ['Age_band_of_driver', 'Driving_experience', 'Casualty_severity', 'Accident_severity']

one_hot_encode_features = [
    'Sex_of_driver', 'Vehicle_driver_relation', 'Type_of_vehicle', 'Owner_of_vehicle',
    'Area_accident_occured', 'Lanes_or_Medians', 'Road_allignment', 'Types_of_Junction',
    'Road_surface_type', 'Road_surface_conditions', 'Light_conditions', 'Weather_conditions',
    'Type_of_collision', 'Vehicle_movement', 'Cause_of_accident', 'Time'
]

label_encoder = LabelEncoder()
for feature in label_encode_features:
    df_clean[feature] = label_encoder.fit_transform(df_clean[feature])

df_clean = pd.get_dummies(df_clean, columns=one_hot_encode_features, drop_first=True).astype(int)

print(df_clean.head())
```

```python
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTETomek
from sklearn.datasets import make_classification
from collections import Counter

X = df_clean.drop(columns=['Accident_severity']) 
y = df_clean['Accident_severity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print('Original dataset shape:', Counter(y_train))

smt = SMOTETomek(random_state=42)
X_train_res, y_train_res = smt.fit_resample(X_train, y_train)
print('Resampled dataset shape:', Counter(y_train_res))
```
