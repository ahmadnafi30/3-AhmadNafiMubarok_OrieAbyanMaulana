```python
import pandas as pd
```

```python
df_1 = pd.read_csv('../datasets/RTA_Datasest_Cleaned_V1.csv')
df_2 = pd.read_csv('../datasets/rta_Clean2.csv')
```

```python
columns = [
    "Time",
    "Day_of_week",
    "Sex_of_driver",
    "Educational_level",
    "Vehicle_driver_relation",
    "Driving_experience",
    "Owner_of_vehicle",
    "Service_year_of_vehicle",
    "Defect_of_vehicle",
    "Area_accident_occured",
    "Type_of_collision",
    "Road_surface_type",
    "Light_condition",
    "Weather_conditions",
    "Vehicle_movement",
    "Casualty_class"
]

valid_columns = [col for col in columns if col in df_1.columns]

df_2[valid_columns] = df_1[valid_columns]

print(df_2)
```

```python
columns_to_drop = [col for col in df_2.columns if col not in df_1.columns]

df_2.drop(columns=columns_to_drop, inplace=True)
print(df_2)
```

```python
df_2.to_csv('../datasets/cleaned(3).csv', index = False)
```
